```python
##Step 1: Extract Text from PDF

#Extract Text from PDF
#using the PyMuPDF (fitz) library

import fitz  #PyMuPDF

#defining function to extract 1st pg
#of pdf and convert to txt
def pdf_to_text_first_page(pdf_path):
    
    # Open the PDF file
    doc = fitz.open(pdf_path)
    
    # Load the first page
    #page indexing in PyMuPDF starts from 0
    #so 0 refers to the first page
    first_page = doc.load_page(0)
    
    # Extract text from the first page
    text = first_page.get_text()
    return text
```

```python
pdf_path = '/Users/jenniferbadger/Dropbox/AI_course/NLP project/joe_20141014_0238_0001.pdf'
first_page_text = pdf_to_text_first_page(pdf_path)
print(first_page_text)
```

```python
##Step 2: Save the Extracted Text to a TXT File

#defining function to save text file
#This function takes two arguments, text and filename.

def save_text_to_file(text, filename):
    #opens the file specified by filename in write mode with UTF-8 encoding.
    with open(filename, 'w', encoding='utf-8') as file:
        #writes the text to the file
        #file is automatically closed when the block inside with is exited
        file.write(text)
        
txt_filename = 'first_page_LeLoi.txt'
save_text_to_file(first_page_text, txt_filename)
```

```python
##Step 3: Translate Text from French to English
#Use the Googletrans library to translate the text from French to English
#googletrans is a Python library that provides a simple interface 
#to Google Translate’s API
from googletrans import Translator

#text: The input text to be translated.
#src_language: The source language code (default is 'fr' for French).
#dest_language: The destination language code (default is 'en' for English).

def translate_text(text, src_language='fr', dest_language='en'):
    #This creates an instance of the Translator class.
    translator = Translator()
    #The translate method of the Translator object 
    #returns a Translated object
    translated = translator.translate(text, src=src_language, dest=dest_language)
    return translated.text
```

```python
first_pg_translated_text = translate_text(first_page_text)
print(first_pg_translated_text)
```

```python
##Step 4: Save the Translated Text to a file

en_txt_filename = 'first_page_LeLoi_en.txt'
save_text_to_file(first_pg_translated_text, en_txt_filename)
```

```python
##Step 5: Split the Translated Text into Sentences using NLTK
#NLTK = Natural Language Toolkit

import nltk
#nltk.download('punkt')
from nltk.tokenize import sent_tokenize

#punkt tokenizer is a pre-trained model 
#that can identify sentence boundaries in a text.
```

```python
#defining the function to split text

def split_text_into_sentences(text):
    #split text into sentences
    sentences = sent_tokenize(text)
    return sentences
```

```python
test_text = "This is a test sentence. Will it work? If so, let's see how it works!"

test_sentences = split_text_into_sentences(test_text)
for sentence in test_sentences:
    print(sentence)
    
#every sentence is a separate line    
```

```python
# Combo of above scripts but in try/except and if/else

pdf_path = '/Users/jenniferbadger/Dropbox/AI_course/NLP project/joe_20141014_0238_0001.pdf'

def pdf_to_text_third_page(pdf_path):
    
    # Open the PDF file
    doc = fitz.open(pdf_path)
    
    # Load the first page
    #page indexing in PyMuPDF starts from 0
    #so 0 refers to the first page
    third_page = doc.load_page(2)
    
    # Extract text from the first page
    text = third_page.get_text()
    return text



# Step 1: Extract Text from PDF
try:
    third_page_text = pdf_to_text_third_page(pdf_path)
    print("Extracted Text from Third Page:\n", third_page_text)
    #debugging
except ValueError as e:
    print(e)
    third_page_text = ""

# Step 2: Save the Extracted Text to a TXT File
if third_page_text:
    txt_filename = 'third_page_LeLoi.txt'
    save_text_to_file(third_page_text, txt_filename)

    # Step 3: Translate Text from French to English
    try:
        third_pg_translated_text = translate_text(third_page_text)
        print("Translated Text:\n", third_pg_translated_text)
        #debugging
    except Exception as e:
        print("Translation error:", e)
else:
    #debugging
    print("No text to translate.")
```
